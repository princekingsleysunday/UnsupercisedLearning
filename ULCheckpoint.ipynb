# HIERARCHICAL 

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage ='complete')


model.fit(new_data)
```

```python
x = new_data[['BALANCE','BALANCE_FREQUENCY', 'CREDIT_LIMIT']]

model.fit(x)
```

```python
clust_label = model.fit_predict(x)

clust_df = pd.DataFrame(clust_label)

clust_df.value_counts()
```

```python
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (data ["PURCHASES_TRX"] , data ["CREDIT_LIMIT"] , c= clust_df[0], s=50)

ax.set_title("Agglomerative Clustering")
ax.set_xlabel("PURCHASES_TRX")
ax.set_ylabel("CREDIT_LIMIT")
plt.colorbar(scatter)

```

```python
from sklearn.cluster import KMeans

model2 = KMeans(n_clusters=5, random_state=0) #cluster by choosing 5 as the number of clusters

x_train = x.loc[:7160]
x_test = x.loc[7160:]

model2.fit(x_train)
```

```python
labels = pd.DataFrame(model2.labels_)

labels.head()
```

```python
new_data2 = pd.concat([data.loc[:7160],labels], axis=1)

new_data.iloc[5].rename({
    0: 'Cluster Labels'
})
```
